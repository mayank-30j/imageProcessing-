```python
# First we will import the libraries

import cv2 
import numpy as np
```

```python
# Now we are going to read a video into cap variable

cap = cv2.VideoCapture('photos/vtest.avi')

# Creating a object 'out' of class 'VideoWriter' for saving the video to a new file.

fourcc = cv2.VideoWriter_fourcc('X','V','I','D')
out = cv2.VideoWriter('output.avi', fourcc, 20,(640,480)) 
```

```python
while cap.isOpened():
    success, frame = cap.read() # Reads the video 
    
    if success == True:         
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # Prints the weight of video
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # Prints the height of video
        
        # Putting some text on the video
        
        font = cv2.FONT_HERSHEY_SIMPLEX
        frame = cv2.putText(frame,'Walking on the street',(100,200),font,2,(0,232,232),2,cv2.LINE_4)
        
        out.write(frame)
        
        # gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # Changes the color of video to grey
        
        cv2.imshow("Just a screen recording",frame) 
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break


cap.release()  
out.release()
cv2.destroyAllWindows()
```
