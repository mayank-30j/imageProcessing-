```python
# First we will import the libraries

import cv2 
import numpy as np
```

```python
# Now we are going to import a video into 'cap' variable

cap = cv2.VideoCapture(0)

# Creating a object 'out' of class 'VideoWriter' for saving the video to a file.

fourcc = cv2.VideoWriter_fourcc('X','V','I','D')
out = cv2.VideoWriter('output.avi', fourcc, 20,(640,480))
```

```python
while cap.isOpened():
    success, frame = cap.read() # Reads the video 
    
    if success == True:         
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # Prints the weight of video
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # Prints the height of video
        
        out.write(frame)
        
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # Changes the color of video to grey
        
        cv2.imshow("Just a screen recording", gray)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break


cap.release()  
out.release()
cv2.destroyAllWindows()
```
